#1. HDI vs TFR Scatter Plot & Regression

```
#Creates HDI analysis dataframe which excludes missing values in the dataframe
dfHDI = df[df['HDI'].notna()]
```

```
#Creating scatter plot of HDI vs TFR  
import pylab 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import plotly.express as px

fig = px.scatter(dfHDI, x = 'HDI', y = 'TFR', color = 'Year',
              title = 'HDI vs TFR ', hover_name = 'Country')

fig.show()
```

```
#linear regression analysis with statsmodels  
import statsmodels.api as sm 

X = dfHDI["HDI"]
Y = dfHDI["TFR"]
X = sm.add_constant(X) # adding an intercept to model


#model = sm.OLS(Y, X).fit()
model = sm.OLS(Y, X.astype(float)).fit()
 
print_model = model.summary()
print(print_model)
```

# 2. Female Education vs TFR Scatterplot & Regression

# Multivariable Regression (IVs: Ed, LFPR)

```
# First check for assumption of no multicolinearity between these 2 variables using VIF
#source: https://www.geeksforgeeks.org/detecting-multicollinearity-with-vif-python/
```

```
df = df.dropna()
```

```
from statsmodels.stats.outliers_influence import variance_inflation_factor 
  
  
# the independent variables set 
X = df[['Females in secondary education (%)','Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)']] 
  
# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                          for i in range(len(X.columns))] 
  
print(vif_data)
```

Hence VIF is large. Some literature says VIF greater than 10 is cause for concern, while other authors more cautious and say VIF greater than 2.5 . Before we proceed with the multiregression analysis we will look at the scatterplot of female education vs female LFPR to assess colinearity

```
#Scatterplot creation

fig = px.scatter(df, x = 'Females in secondary education (%)', y = 'Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)', color = 'Year',
              title = 'Female LFPR vs TFR', hover_name = 'Country')
fig.show()
```

May justify that from graph, variables do not look linear, hence may proceed with multiregression analysis. 

```
# Multi Regression Analysis 
df = df.dropna()

#MVR with education & LFPR, assume no constant, source: https://towardsdatascience.com/simple-and-multiple-linear-regression-in-python-c928425168f9
import statsmodels.api as sm

X = df[['Females in secondary education (%)','Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)']] 
y = df['TFR']


model = sm.OLS(y, X).fit()

# Print out the statistics
model.summary()
```

```
#MVR with education & LFPR, assume constant source: https://towardsdatascience.com/simple-and-multiple-linear-regression-in-python-c928425168f9

X = df[['Females in secondary education (%)','Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)']] 
y = df['TFR']
X = sm.add_constant(X) # adding an intercept to model

model = sm.OLS(y, X).fit() 

# Print out the statistics
model.summary()
```
