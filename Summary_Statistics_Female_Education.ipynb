# Import Commands and Data Uploading 



```
import pandas as pd 
import pylab
import matplotlib.pyplot as plt
```

```
#Uploading the Cleaned Female Education Dataset 

from google.colab import files
uploaded = files.upload()


```

```
data_path = "/content/2. Female:Education:Cleaned:Country.csv"
female_edu= pd.read_csv(data_path, index_col=0)
female_edu
```

# Cleaning the dataframe

```
#cleaning Time period by converting the string to integer 
female_edu['Time Period'] = female_edu['Time Period'].fillna(0).astype(int)
female_edu
```

```
#narrowing the timeframe down as most values lie between 1975-2014

female_edu= female_edu[female_edu['Time Period'] <= 2014]
female_edu= female_edu[female_edu['Time Period'] >= 1975]
female_edu.reset_index(drop=True)
female_edu
```

# Reshaping the data

```
#reshaping the dataframe 

female_edu_reshaped = female_edu.pivot_table(index='Reference Area', columns='Time Period', values='Observation Value')
female_edu_reshaped.head()

```

```
#saving new file 

female_edu_reshaped.to_csv('Female Education Reshaped.csv')
```

# Creating Dataframe with only African countries 

```
#creating a new dataframe with only African countries
from google.colab import files
uploaded = files.upload()
```

```
africa = "/content/Africa_FULL_MERGE.csv"
female_edu_africa = pd.read_csv(africa, index_col=0, encoding='latin1')

#narrowing table down to Female education

female_edu_africa = female_edu_africa[['Country', 'Region', 'Year','Females in secondary education (%)']]
female_edu_africa.head()
```

```
#narrowing down timeframe as Female Education data recorded mainly between 1975-2014
female_edu_africa= female_edu_africa[female_edu_africa['Year'] <= 2014]
female_edu_africa = female_edu_africa[female_edu_africa['Year'] >= 1975]
female_edu_africa.reset_index(drop=True)
female_edu_africa
```

```
#Reshaping the Data 

female_edu_africa_new = female_edu_africa.pivot_table(index=['Country', 'Region'], columns='Year', values='Females in secondary education (%)')
female_edu_africa_new.head()
```

```
#saving new file

female_edu_africa_new.to_csv('Female Education Africa.csv')
```
