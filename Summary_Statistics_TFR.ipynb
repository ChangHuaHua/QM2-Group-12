# VISUALIZATION: line plot africa v world average 

```
#Select row for median and alter to data frame (World)
meanw = world.loc['mean']
world_mean=meanw.to_frame()
world_mean.head()
```

```
meana = africa.loc['mean']
africa_mean = meana.to_frame()
africa_mean.head()
```

```
#Merge mean data into one dataframe to aid plotting 
means = world_mean.merge(africa_mean, left_index = True, right_index=True)
means.rename(columns={'mean_x':'World','mean_y':'Africa'}, inplace=True)
means.reset_index(inplace=True)
means.rename(columns={'index':'Year'}, inplace=True)

means.head()
```

```
ax = means.plot( x= 'Year', y= 'World', kind='line', c = 'blue', label = 'World')
means.plot(x='Year', y= 'Africa', kind='line', c = 'orange', ax=ax, label = 'Africa')
ax.legend(loc='upper right')
plt.title('Average World and African Total Fertilty Rates from 1960-2018')
plt.ylabel('Total Ferilty Rate')
```

```
#alter to long form data bc more compatible https://altair-viz.github.io/user_guide/data.html?highlight=wide#long-form-vs-wide-form-data 
means= means.melt('Year', var_name='Region', value_vars=list(means.columns)[1:3])
means
```

```
#Plotly visualization
import plotly.express as px 

fig = px.line(means, x = 'Year', y = 'value',
             color = 'Region', line_group = 'Region',
             title = 'Figure 1: Average World and Average African Total Fertilty Rates (TFR) from 1960-2018',
             labels={"value": "Average TFR"})
 
fig.for_each_trace(lambda t: t.update(name=t.name.replace("Region=", "")))


#layout code format from https://plotly.com/python/setting-graph-size/ 
fig.update_layout(
    autosize=False,
    width=800,
    height=600,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ))


fig.show()
```

Average TFR in Africa consistently been significantly higher than that of the average world TFR. Hence justifies our focus on this content, and supports existing literature.
