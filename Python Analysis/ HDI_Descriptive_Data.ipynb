# **Creating Dataframe with only African countries**


```
#Upload Africa_FULL_MERGE data
from google.colab import files
uploaded = files.upload()
```

```
data2 = "Africa_FULL_MERGE.csv"
dataHDI_Africa = pd.read_csv(data2, index_col=0, encoding='latin1')
dataHDI_Africa.head()
```

```
#narrowing table down to HDI data
dataHDI_Africa = dataHDI_Africa[['Country', 'Year', 'Region', 'HDI','HDI Rank (2018)']]
dataHDI_Africa.head()
```

```
#narrowing down timeframe as HDI data recorded mainly between 1975-2018
dataHDI_Africa = dataHDI_Africa[dataHDI_Africa['Year'] <= 2018]
dataHDI_Africa = dataHDI_Africa[dataHDI_Africa['Year'] >= 1975]
dataHDI_Africa.reset_index(drop=True)
dataHDI_Africa
```

```
#saving new file
dataHDI_Africa.to_csv('HDI Africa.csv')
```

# **Summary Statistics: Africa HDI Data**

```
#Again must convert HDI values of object types to float type in order to get summary stats
dataHDI_Africa = pd.DataFrame(dataHDI_Africa)


dataHDI_Africa['HDI']=pd.to_numeric(dataHDI_Africa['HDI'], errors='coerce')
dataHDI_Africa['HDI Rank (2018)']=pd.to_numeric(dataHDI_Africa['HDI Rank (2018)'], errors='coerce')

dataHDI_Africa.dtypes
```

```
dataHDI_Africa.describe()
```

```
#Most recently in 2018, Mean World HDI is 0.713, while for the African continent it is 0.551

dataHDI_2018 = dataHDI_Africa[dataHDI_Africa['Year'] == 2018]
dataHDI_2018 = dataHDI_2018.reset_index(drop=True)
dataHDI_2018 = dataHDI_2018[['Country', 'Year', 'HDI', 'HDI Rank (2018)']]
dataHDI_2018
```
