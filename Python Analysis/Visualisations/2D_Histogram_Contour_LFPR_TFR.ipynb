# Import commands and data uploading

```
#Import commands

import pandas as pd
import plotly.express as px
```

```
#Uploading the Full Merge Africa Dataset 
from google.colab import files
uploaded = files.upload()
```

```
africa = pd.read_csv('/content/Africa_FULL_MERGE.csv', index_col=0)
africa
```

# Narrowing down the dataframe

```
#Narrowing down the dataframe to TFR, HDI and Female Labour Participation

africa = africa[['Country', 'Region', 'Year','TFR','HDI', 'Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)']]
africa

```

```
#Narrowing down the dataframe to 1990-2017 time period
africa_new = africa[africa['Year'] >= 1990]
africa.reset_index(drop=True)
africa_new
```

```
africa_new2 = africa_new[africa_new['Year'] <= 2017]
africa_new.reset_index(drop=True)
africa_new2
```

```
#Saving the file
africa_new2.to_csv('LFPR in Africa between 1990 and 2017')
```

# Plotting the 2D Histogram Contour figure 

```
#Plotting the figure to clarify the scatter plot
fig_africa = px.density_contour(africa_new2, x="Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)", y="TFR")
fig_africa.update_traces(contours_coloring="fill", contours_showlabels = True)
fig_africa.show()
```
