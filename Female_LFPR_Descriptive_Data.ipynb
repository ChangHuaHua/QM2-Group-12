**Creating Dataframe with only African countries**

```
FemaleLFPR_Africa = FemaleLFPR_Africa[['Country', 'Year', 'Region', 'Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)']]
FemaleLFPR_Africa.head()
```

```
FemaleLFPR_Africa = FemaleLFPR_Africa[FemaleLFPR_Africa['Year'] <= 2017]
FemaleLFPR_Africa = FemaleLFPR_Africa[FemaleLFPR_Africa['Year'] >= 1990]
FemaleLFPR_Africa.head()
```
